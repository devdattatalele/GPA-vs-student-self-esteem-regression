## 1. Importing Libraries and Loading Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/content/student_data (2).csv')
```

```python
print(data.head())
print(data.info())
```

### Descriptive Statistics

```python
print(data.describe())
```

### Data Visualization

```python
# Histogram of GPA
plt.figure(figsize=(10, 6))
sns.histplot(data['gpa'], bins=20, kde=True)
plt.title('Distribution of GPA')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of GPA vs. Study Hours
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='study_hours', y='gpa')
plt.title('GPA vs. self esteem')
plt.xlabel('self esteem')
plt.ylabel('GPA')
plt.show()
```

```python
print(data.isnull().sum())
```

```python
# Box plot of Study Hours
plt.figure(figsize=(8, 5))
sns.boxplot(data['study_hours'])
plt.title('Box Plot of selfesteem')
plt.xlabel('Self_esteem')
plt.show()
```

```python
# Remove outliers using IQR method
Q1 = data['study_hours'].quantile(0.25)
Q3 = data['study_hours'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data = data[(data['study_hours'] >= lower_bound) & (data['study_hours'] <= upper_bound)]
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.lmplot(data=data, x='study_hours', y='gpa')
plt.title('GPA vs. Self_esteem (Regression)')
plt.xlabel('Self_esteem')
plt.ylabel('GPA')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(data['gpa'], label='GPA')
sns.kdeplot(data['study_hours'], label='Self_esteem')
plt.title('Distribution of GPA and Self_esteem')
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()
plt.show()
```
